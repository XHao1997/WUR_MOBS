```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("temp_real.csv",usecols=['X','Y'])
```

```python
plt.scatter(df['X'],df['Y'])
```

```python
interpolate_days = np.arange(0,252,0.1)
```

```python
y_int = np.interp(interpolate_days,df['X'],df['Y'])
```

```python
plt.scatter(interpolate_days,y_int)
```

```python
interpolate_days.shape
```

```python
m_grass = np.array([])
for i in range(interpolate_days.shape[0]):
    m_grass.insert(np.trapz(y_int[:i], interpolate_days[:i]))
```

```python
plt.scatter(interpolate_days,m_grass,s=1)
```
